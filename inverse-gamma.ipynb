```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
trueTheta=42.0
```

Few samples:

```python
n0 = 3
x0 = np.random.normal(loc=0.0, scale=np.sqrt(trueTheta), size=n0)
```

More samples:

```python
n1 = 100
x1 = np.random.normal(loc=0.0, scale=np.sqrt(trueTheta), size=n1)
```

Lots of samples:

```python
n2 = 500
x2 = np.random.normal(loc=0.0, scale=np.sqrt(trueTheta), size=n2)
```

Plot posterior:

```python
theta=np.linspace(0,100,1000)

dist=st.invgamma.pdf(theta, a=n0/2, scale=np.sum(x0**2)/2)
plt.plot(theta, dist, label="n = %d" % n0)

dist=st.invgamma.pdf(theta, a=n1/2, scale=np.sum(x1**2)/2)
plt.plot(theta, dist, label="n = %d" % n1)

dist=st.invgamma.pdf(theta, a=n2/2, scale=np.sum(x2**2)/2)
plt.plot(theta, dist, label="n = %d" % n2)

plt.legend()
```
